# ML_Project_01 : Increase sales of an online music store

### Import & Read the dataset

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv('music.csv')
df
```

### Split the data (input & output sets)

```python
X = df.drop(columns=['genre'])     #this is keep the original df and creates a copy
#input set
X
```

```python
y = df['genre']        #taking the column we need for predictions
# output set
y
```

### Create a model

```python
# create a new instance (model) from DecisionTreeClassifier class
model = DecisionTreeClassifier()
```

This model needs to be trained using the data to find patterns

### Train the model

```python
# using fit(input, output)
model.fit(X, y)
```

### Ask model to make predictions

```python
# askig to make 2 predictions at once
predictions = model.predict( [ [21,1], [20,0] ])
predictions
```

### Check the accuracy of the predictions

```python
from sklearn.model_selection import train_test_split
```

```python
# call the train_test_split function
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2 )
```

```python
# create a model to check accuracy
acc_model = DecisionTreeClassifier()
```

```python
acc_model.fit(X_train, y_train)
# test using 
acc_predictions = acc_model.predict(X_test)
```

```python
# to calculate the accuracy, compare the predictions with the actual values we have in the output set for testing

# call the accuracy_score function
score = accuracy_score( y_test, acc_predictions )
score
```
